```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
investment = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50]).reshape(-1, 1)
sales_growth = np.array([25, 30, 40, 45, 50, 60, 65, 70, 80])

model = LinearRegression()
model.fit(investment,sales_growth)

print(f'Slope (m): {model.coef_[0]:.2f}')
print(f"Intercept (b): {model.intercept_:.2f}") 

equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f'Line equation: {equation}')

predicted_sales = model.predict([[60]]) 
print(f'Predicted sales growth for 60k investment: {predicted_sales[0]:.2f}')

required_investment = (100-model.intercept_) / model.coef_[0]
print(f'Required investment for 100k sales growth:{required_investment:.2f}')

plt.figure(figsize=(10,6))
plt.scatter(investment , sales_growth , color = 'blue',label='Data points')
plt.plot(investment , model.predict(investment),color = 'red',label='Regression line')
plt.scatter([[60]],[predicted_sales],color='green',s=100,label='Prediction for 60k')

plt.xlabel('Advertising Investment (1000)')
plt.ylabel('Sales growth(1000)')
plt.title('Liner Regression - Advertising vs Sales growth')
plt.grid(True)
plt.legend()
plt.text(15,80,equation,fontsize=12)
plt.show()
```
